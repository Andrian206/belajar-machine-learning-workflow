Checking of missing values

```python
missing_values = train.isnull().sum()
missing_values[missing_values > 0]
```

Separated features by missing value (Numeric)

```python
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values >= 1000].index
```

Fill in the missing value (Numeric)

```python
numeric_features = train[less].select_dtypes(include=['number']).columns
train[numeric_features] = train[numeric_features].fillna(train[numeric_features].median())
```

Fill in the missing value (Category)

```python
categorical_feature = train[less].select_dtypes(include=['object', 'str']).columns

for column in categorical_feature:
    train[column] = train[column].fillna(train[column].mode()[0])
```

Deleting columns with too many missing values (over)

```python
df = train.drop(columns=over)
```

Verify the datasets

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

Checking of the outliers

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

selected_features = ['OverallQual', 'OverallCond']

for feature in selected_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

Identify outliers by IQR

```python
Q1 = df[selected_features].quantile(0.25)
Q3 = df[selected_features].quantile(0.75)
IQR = Q3 - Q1
```

Delete outliers by IQR

```python
condition = ~((df[selected_features] < (Q1 - 1.5 * IQR)) | (df[selected_features] > (Q3 + 1.5 * IQR))).any(axis=1)
df_filtered_selected = df.loc[condition, selected_features]

categorical_feature = train[less].select_dtypes(include=['object', 'str']).columns
df = pd.concat([df_filtered_selected, df.loc[condition, categorical_feature]], axis=1)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

for feature in selected_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature} (After Cleaning)')
    plt.show()
```
